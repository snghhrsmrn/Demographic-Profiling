```python
data_2013 = pd.read_excel('AC23_2013.xlsx')
data_2015 = pd.read_excel('AC23_2015.xlsx')
```

```python
data_2013_part2 = data_2013.iloc[1, 8:10]
data_2015_part2 = data_2015.iloc[1, 11:13]
total_voters_2013 = np.sum(data_2013_part2.values)
total_voters_2015 = np.sum(data_2015_part2.values)
total_population = len(data['Name'])

total_voters_2015
```

# voter turnout 2013

```python
voter_turnout_2013 = (total_voters_2013/total_population)*100
voter_turnout_2013
```

# voter turnout 2015

```python
voter_turnout_2015 = (total_voters_2015/total_population)*100
voter_turnout_2015
```

```python
Age_group = data.groupby(['Age_group'])['Age'].count()
```

# 68% of voters who voted in 2013

```python
age_group_count_2013 = ((Age_group*voter_turnout_2013)/100).astype('int')
age_group_count_2013
```

# 77% of voters who voted in 2015

```python
age_group_count_2015 = ((Age_group*voter_turnout_2015)/100).astype('int')
age_group_count_2015
```

# percentage of votes given in 2013 category

```python
perOfEachCategory_2013 = (age_group_count_2013/total_voters_2013)
perOfEachCategory_2013
```

# percentage of votes given in 2015 category

```python
perOfEachCategory_2015 = (age_group_count_2015/total_voters_2015)*100
perOfEachCategory_2015
```

# percentage of partywise votes and converting into DataFrame

```python
party_data_2013 = data_2013.iloc[1, 1:4]/total_voters_2013
# party_data_2013 = pd.DataFrame(data = [party_data_2013])
party_data_2015 = data_2015.iloc[1, 1:4]/total_voters_2015
# party_data_2015 = pd.DataFrame(data = [party_data_2015])
```

```python
per_voters_2013 = pd.DataFrame(index = list(perOfEachCategory_2013.index), columns=['Congress', 'BJP', 'AAP'])
per_voters_2015 = pd.DataFrame(index = list(perOfEachCategory_2015.index), columns=['Congress', 'BJP', 'AAP'])
```

```python
list(party_data_2015.values)[0]
```

```python
for i in range(len(perOfEachCategory_2013)):
    for j in range(len(party_data_2013)):
        per_voters_2013.iloc[i,j] = (perOfEachCategory_2013[i]*list(party_data_2013.values)[j])*100
        
for i in range(len(perOfEachCategory_2015)):
    for j in range(len(party_data_2015)):
        per_voters_2015.iloc[i,j] = (perOfEachCategory_2015[i]*list(party_data_2015.values)[j])*100
```
